#Task 6: Prediction Using Decision Tree


Decision Tree is a Supervised Machine Learning algorithm. In the given problem statement, we have to create a Decision Tree Classifier and visualize it graphically so that whenever we feed any new data to the classifer, it would be able to predict the right class.

#Author: Shreyans Upadhyay

#Step 1: Importing the necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

#Step 2: Loading the dataset

```python
data=load_iris()
```

```python
#check the rows and columns of the dataset
data.data.shape
```

```python
print("Classes to predict:", data.target_names)
```

```python
print("Features:", data.feature_names)
```

```python
x=data.data
y=data.target
```

```python
display(x.shape, y.shape)
```

#Step 3: Building the Model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest=train_test_split(x, y, random_state=50, test_size=0.30)
```

```python
clf=DecisionTreeClassifier()
clf.fit(xtrain, ytrain)
```

```python
y_pred=clf.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy on train data:", accuracy_score(y_true=ytrain, y_pred=clf.predict(xtrain)))
print("Accuracy on test data:", accuracy_score(y_true=ytest, y_pred=y_pred))
```

```python
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
```

```python
dot_data=StringIO()
export_graphviz(clf, out_file=dot_data, filled=True, rounded=True, special_characters=True, 
                feature_names=data.feature_names, class_names=data.target_names)
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
